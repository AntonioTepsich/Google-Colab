FASE 2: entrenamiento

uso 80% para entrenar que es lo optimo y 20% para testear los resultados

```python
datos_entrenamiento=datos_maraton.sample(frac=0.8,random_state=0)
datos_test=datos_maraton.drop(datos_entrenamiento.index)
```

```python
datos_entrenamiento
```

separo a los datos del tiempo final de la maraton para compararlos despues

```python
etiquetas_entrenamiento=datos_entrenamiento.pop("MarathonTime")
etiquetas_test=datos_test.pop("MarathonTime")
```

```python
etiquetas_entrenamiento
```

```python
datos_entrenamiento
```

entreno con una regresion lineal (manera mas facil de hacerlo)

```python
from sklearn.linear_model import LinearRegression
modelo= LinearRegression()
modelo.fit(datos_entrenamiento,etiquetas_entrenamiento)
```

esta es la prediccion del 20% restante

```python
predicciones=modelo.predict(datos_test)
predicciones
```

calculo el error que hubo con los valores oficiales de ese 20%

```python
import numpy as np
from sklearn.metrics import mean_squared_error
error=np.sqrt(mean_squared_error(etiquetas_test,predicciones))
print("Error porcentual : %f" %(error*100))
```

11% de error para este tipo de problemas es aceptable
